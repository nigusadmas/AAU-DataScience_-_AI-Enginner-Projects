```python
# Section D: Data Cleaning
# Exercise 18: Missing Values Detection (0.6 marks)
# Create a DataFrame containing missing values.
import pandas as pd
import numpy as np
df = pd.DataFrame({
    "Student_ID": [101, 102, 103, 104, 105, 106],
    "Name": ["Abebe", "Sara", "John", None, "Mekdes", "Daniel"],
    "Age": [20, 21, None, 22, 19, 23],
    "Score": [85, None, 78, 90, None, 88],
    "Department": ["IT", "CS", None, "IT", "SE", "CS"],
    "Attendance": [95.5, 88.0, 92.5, None, 85.0, None]
})
print(df)
# Write code to determine:
# Number of missing values in each column.
print(df.isnull().sum())
# Total missing values in the dataset.
print("========================")
print("Total missing Value:",df.isnull().sum().sum())
```

```python
# Exercise 19: Handling Missing Values (0.7 marks)
# Replace missing numerical values using:

# ==========================
# Mean replacement
# ==========================

df_mean = df.copy()

# Fill missing numerical values with mean
df_mean["Age"] = df_mean["Age"].fillna(df_mean["Age"].mean())
df_mean["Score"] = df_mean["Score"].fillna(df_mean["Score"].mean())
df_mean["Attendance"] = df_mean["Attendance"].fillna(
    df_mean["Attendance"].mean()
)

print("DataFrame after Mean replacement:")
print(df_mean)


print("========================================")


# ==========================
# Median replacement
# ==========================

df_median = df.copy()

# Fill missing numerical values with median
df_median["Age"] = df_median["Age"].fillna(df_median["Age"].median())
df_median["Score"] = df_median["Score"].fillna(df_median["Score"].median())
df_median["Attendance"] = df_median["Attendance"].fillna(
    df_median["Attendance"].median()
)

print("DataFrame after Median replacement:")
print(df_median)


print("========================================")


# Compare Mean vs Median results

comparison = pd.DataFrame({
    "Mean Replacement": [
        df["Age"].mean(),
        df["Score"].mean(),
        df["Attendance"].mean()
    ],
    "Median Replacement": [
        df["Age"].median(),
        df["Score"].median(),
        df["Attendance"].median()
    ]
})

print("Comparison:")
print(comparison)
```

```python
# Section E: Integrated Data Wrangling
# Exercise 23: Cleaning a Customer Dataset (1 mark)
import pandas as pd
import numpy as np

# Create dataset with:
# - Missing values
# - Duplicate rows
# - Inconsistent text

df = pd.DataFrame({
    "Customer_ID": [101, 102, 103, 104, 104, 105],
    "Name": ["John", "mary", "DAVID", "Sara", "Sara", "Mike"],
    "City": ["Addis Ababa", "addis ababa", "ADDIS ABABA", None, "Addis Ababa", "Bahir Dar"],
    "Age": [25, 30, np.nan, 28, 28, 35]
})

print("Original Dataset:")
print(df)


# -------------------------------
# Cleaning Process
# -------------------------------

# 1. Remove duplicate rows
df = df.drop_duplicates()


# 2. Handle missing values
# Fill missing city with "Unknown"
df["City"] = df["City"].fillna("Unknown")

# Fill missing age with mean age
df["Age"] = df["Age"].fillna(df["Age"].mean())


# 3. Fix inconsistent text
# Convert names to proper case
df["Name"] = df["Name"].str.title()

# Convert city names to proper case
df["City"] = df["City"].str.title()


print("\nCleaned Dataset:")
print(df)
```

```python
import pandas as pd

# Sales dataset
sales = pd.Series([100, 120, 130, 150, 160, 170, 180, 190, 200, 500])

# Calculate Q1 and Q3
Q1 = sales.quantile(0.25)
Q3 = sales.quantile(0.75)

# Calculate IQR
IQR = Q3 - Q1

# Calculate bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = sales[(sales < lower_bound) | (sales > upper_bound)]

# Display results
print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)

print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)

print("\nOutlier Values:")
print(outliers)
```

```python
import pandas as pd
import numpy as np

# ---------------------------------------
# Create dataset with data problems
# ---------------------------------------

df = pd.DataFrame({
    "Customer_ID": [101, 102, 103, 104, 104, 105, 106],
    "Name": ["john", "MARY", "David", "sara", "sara", "MIKE", "Alex"],
    "City": ["addis ababa", "ADDIS ABABA", None, "Bahir dar", "Bahir dar", "dire dawa", "Addis Ababa"],
    "Sales": [500, 600, 700, 650, 650, 550, 10000]
})

print("Original Dataset:")
print(df)


# ---------------------------------------
# 1. Detect Missing Values
# ---------------------------------------

print("\nMissing Values:")
print(df.isnull().sum())


# ---------------------------------------
# 2. Fill Missing Values
# ---------------------------------------

# Fill missing city
df["City"] = df["City"].fillna("Unknown")


# ---------------------------------------
# 3. Remove Duplicate Records
# ---------------------------------------

df = df.drop_duplicates()


# ---------------------------------------
# 4. Standardize Text Values
# ---------------------------------------

df["Name"] = df["Name"].str.title()
df["City"] = df["City"].str.title()


# ---------------------------------------
# 5. Detect Outliers Using IQR
# ---------------------------------------

Q1 = df["Sales"].quantile(0.25)
Q3 = df["Sales"].quantile(0.75)

IQR = Q3 - Q1


lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR


print("\nLower Bound:", lower_bound)
print("Upper Bound:", upper_bound)


# Display outliers
outliers = df[
    (df["Sales"] < lower_bound) |
    (df["Sales"] > upper_bound)
]

print("\nOutliers:")
print(outliers)


# ---------------------------------------
# 6. Remove/Capping Outliers
# ---------------------------------------

# Cap extreme values instead of deleting
df["Sales"] = np.where(
    df["Sales"] > upper_bound,
    upper_bound,
    df["Sales"]
)


# ---------------------------------------
# 7. Display Final Clean Dataset
# ---------------------------------------

print("\nFinal Cleaned Dataset:")
print(df)
```
