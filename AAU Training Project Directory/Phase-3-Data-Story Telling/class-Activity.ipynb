```python
# Questions:which day been the maximum sales using graph
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

sales = {
    "day": ["Mon", "Tue", "Wednes", "Thurs",
            "Fri", "Satur", "Sun"],
    "cups_sold": [120, 135, 110, 145, 160, 210, 190]
}

ax.bar(sales["day"], sales["cups_sold"], width=0.7)

ax.set_title("Coffee Sales by Day")
ax.set_xlabel("Day")
ax.set_ylabel("Cups Sold")

plt.show()
```
