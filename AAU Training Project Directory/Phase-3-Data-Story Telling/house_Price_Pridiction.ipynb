```python
import pandas as pd 

import numpy as np

from sklearn.linear_model import LinearRegression

print("This is the independent value(Features)")
X=[[50],[60],[70],[80],[90],[100],[120]]

# print(X)

print("This is the dependent varaibles on X(Target Value(Sample data that are assigned to X))")
y=[[500],[600],[700],[800],[900],[1000],[1200]]


# print(y)

model=LinearRegression()



model.fit(X,y)

print("The machine is finished studying the data")

new_House=np.array([[130],[300]])


predictions=model.predict(new_House)
predictions = predictions.flatten()

# 7. ውጤቱን ማሳየት
print(f"ማሽኑ የገመተው የ 130 ካሬ ቤት ዋጋ፦ ETB {predictions[0]:,.2f} ሺህ ብር")
print(f"ማሽኑ የገመተው የ 300 ካሬ ቤት ዋጋ፦ ETB {predictions[1]:,.2f} ሺህ ብር")

```
