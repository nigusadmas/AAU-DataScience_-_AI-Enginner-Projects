```python
# Exercise 1: Monthly Sales Trend

import matplotlib.pyplot as plt

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
sales = [120, 150, 170, 160, 190]

plt.plot(months,sales ,marker="s" ,label="Sales" , markersize=8,color="green",linestyle="dashed")

plt.title("Monthly Sales.")

plt.xlabel("months")

plt.ylabel("Sales amount")

plt.legend()

plt.show()
```

```python
# Exercise 2: Student Scores

import matplotlib.pyplot as plt

subjects = ['Math', 'Physics', 'Chemistry', 'Biology']
scores = [85, 78, 92, 88]

plt.plot( subjects,scores ,marker="*", markersize=10 , linewidth=6,color="red" ,markerfacecolor="yellow" ,label="Scores")

plt.title("Exam Scores")

plt.xlabel("subjects")
plt.ylabel("Exam Scores")

plt.legend()

plt.show()
```

```python
# Exercise 3: Company Revenue
import matplotlib.pyplot as plt

years = [2021, 2022, 2023, 2024]
revenue = [500, 700, 850, 1000]

plt.plot(years ,revenue ,color="blue" , linestyle="dotted" , marker="d" ,label="Revenue")

plt.xlabel("Years(Y)")
plt.ylabel("Revenue In Year (R)")

plt.title("Revenue Of The Product In A Year")

plt.legend()

plt.show()

print("This diagram indicates the revenue of the produt increases Over Year to Year")
```

```python
# Exercise 4: Fruit Supply (Vertical Bar Chart)
import matplotlib.pyplot as plt

# Here is the Exact data to be manipulated
fruits = ['Apple', 'Orange', 'Banana', 'Mango']
counts = [50, 70, 40, 60]
colors = ['red', 'orange', 'yellow', 'green']
bar_colors = ['tab:red', 'tab:blue', 'tab:green', 'tab:orange']
# Use the following two word to set Outlier figure and to assign proper content to a figure
fig,ax=plt.subplots()

ax.bar(fruits,counts, label=colors , color=bar_colors)

ax.set_ylabel("Quantity")

ax.set_title("Fruit Supply")

ax.legend()

plt.show()

```

```python
# Exercise 5: Stacked Bar Chart (Students by Gender)
import numpy as np
import matplotlib.pyplot as plt

departments = ('CS', 'IS', 'SE', 'IT')

students = {
    'Male': np.array([70, 65, 80, 50]),
    'Female': np.array([60, 55, 70, 45])
}
width = 0.75
fig, ax = plt.subplots()
bottom = np.zeros(4)

for gender, counts in students.items():
    bars = ax.bar(
        departments,
        counts,
        width=width,
        bottom=bottom,
        label=gender
    )

    # Add labels inside each bar section
    ax.bar_label(bars, label_type='center')

    bottom += counts

ax.set_title("Students by Department and Gender")
ax.legend()

plt.show()
```
