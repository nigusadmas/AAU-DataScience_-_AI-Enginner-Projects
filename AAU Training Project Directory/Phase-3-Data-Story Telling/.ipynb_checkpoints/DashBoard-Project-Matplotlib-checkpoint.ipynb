```python
import matplotlib.pyplot as plt
import numpy as np


# -------------------------------
# Create Dashboard Layout
# -------------------------------
fig, ax = plt.subplots(2, 2, figsize=(14, 10))


# ==================================================
# 1. Monthly Sales Trend (Line Chart)
# ==================================================

months = [
    "Jan", "Feb", "Mar", "Apr",
    "May", "Jun", "Jul", "Aug",
    "Sep", "Oct", "Nov", "Dec"
]

sales = [
    12000, 15000, 17000, 14000,
    18000, 22000, 25000, 23000,
    27000, 30000, 32000, 35000
]


ax[0,0].plot(months, sales, marker="o" ,markersize=12, markerfacecolor="yellow" , linewidth=4)

ax[0,0].set_title("Monthly Sales Trend")
ax[0,0].set_xlabel("Month")
ax[0,0].set_ylabel("Sales ($)")

ax[0,0].grid(True)



# ==================================================
# 2. Fruit Supply (Vertical Bar Chart)
# ==================================================

fruits = [
    "Apple",
    "Banana",
    "Orange",
    "Mango",
    "Grapes"
]

quantity = [
    500,
    700,
    450,
    600,
    800
]

colors=['red','yellow','green','orange','blue']
bar_color=['tab:red','tab:yellow','tab:green','tab:orange','tab:blue']
bars = ax[0,1].bar(
    fruits,
    quantity,
    label="Fruit Quantity",
    color=bar_color
)

ax[0,1].set_title("Fruit Supply")
ax[0,1].set_xlabel("Fruit")
ax[0,1].set_ylabel("Quantity")

ax[0,1].legend()

ax[0,1].bar_label(
    bars,
    label_type="center"
)


# ==================================================
# 3. Movie Ratings (Horizontal Bar Chart)
# ==================================================

movies = [
    "Inception",
    "Avatar",
    "Titanic",
    "Joker",
    "Interstellar"
]


ratings = [
    8.8,
    7.9,
    7.8,
    8.4,
    8.6
]


bars2 = ax[1,0].barh(
    movies,
    ratings
)


ax[1,0].set_title("Movie Ratings")
ax[1,0].set_xlabel("Rating")


ax[1,0].bar_label(bars2)



# ==================================================
# 4. Student Distribution by Gender
# (Stacked Bar Chart)
# ==================================================

departments = [
    "CS",
    "IS",
    "SE",
    "IT"
]


male = np.array(
    [70,65,80,50]
)


female = np.array(
    [60,55,70,45]
)


# Bottom starts from zero
bottom = np.zeros(4)


ax[1,1].bar(
    departments,
    male,
    label="Male"
)


ax[1,1].bar(
    departments,
    female,
    bottom=male,
    label="Female"
)


ax[1,1].set_title(
    "Student Distribution by Gender"
)

ax[1,1].set_xlabel(
    "Department"
)

ax[1,1].set_ylabel(
    "Students"
)


ax[1,1].legend()



# Adjust spacing
plt.tight_layout()


# Display dashboard
plt.show()
```
