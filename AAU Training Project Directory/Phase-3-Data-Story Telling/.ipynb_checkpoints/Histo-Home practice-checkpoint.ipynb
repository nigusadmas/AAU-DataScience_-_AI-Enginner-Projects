```python
import matplotlib.pyplot as plt
import numpy as np
scores = [
45,50,52,55,57,
60,61,62,65,67,
70,71,72,75,78,
80,82,85,88,90
]

plt.hist(scores , bins=5 , edgecolor="black")

plt.title("Student Score Distribution")

plt.xlabel("scores")

plt.ylabel("Number of Students")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

scores = np.random.normal(
    loc=70,
    scale=10,
    size=1000
)

plt.hist(scores,bins=20,edgecolor='black')

plt.title("Normal Distribution")
plt.show()
```

```python
bills = [
100,120,150,180,
200,220,250,
300,350,400,
500,800,1500,5000
]

plt.hist(bills,bins=10,edgecolor='black')
plt.xlabel("Amount Spended")
plt.ylabel("Number of Persons")
plt.title("Right Skewed Histogram")
plt.show()
```

```python
scores = [
40,60,
70,75,
80,82,84,85,
87,88,89,
90,91,92,93,
95,96,98
]

plt.hist(scores,bins=8)

plt.xlabel("Amount Scored(100%)")
plt.ylabel("Number of Students")
plt.title("Left Skewed Histogram")
plt.show()
```

```python
data = np.random.uniform(
    0,
    100,
    1000
)

plt.hist(data,bins=10)


plt.xlabel("Random Distribution(X)")
plt.ylabel("Random Distribution(Y)")
plt.title("Uniform Distribution Histogram")
plt.show()
```
