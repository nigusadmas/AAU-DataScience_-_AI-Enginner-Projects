```python
# Task2
#####################################################################
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
movies=pd.read_csv("movies_updated.csv")
# movies.head()
# movies.info()
# movies.describe()


# Questions
# 1. Which year has the most movies?
movie_per_year=movies["year"].value_counts().sort_index()
most_movies=movie_per_year.idxmax()
most_year=movie_per_year.max()
print('Year:',most_movies)
print('Value',most_year)
#####################################################################

# 2. Does the number of movies increase over time?
plt.figure(figsize=(10,5))
plt.plot(
    movie_per_year.index,
    movie_per_year.values,
    color="red",
    marker="d",
    linewidth=3,
    markersize=10,
    markerfacecolor="yellow"
    
)
plt.title("Number Of Movies Per Year" , c="blueviolet" , size=20 )
plt.xlabel("Years",c="red" , size=16)
plt.ylabel("Number Of Movies For Each Year",c="orange" , size=16)
plt.show()
#####################################################################

# 3. What trend do you observe?
    # The line plot shows there is a flactuation of values based on the dataset
    # Movie production grows during some periods(1980-1984)
    # Certain years show peaks (highest production)(1985)
    # Some years may show in either increase nor decrease.(1985-2000)
    # After some Year it can be declined (Year>2000)
    # The overall trend can be increasing, decreasing, or fluctuating depending on  dataset.
```

```python
# Task 3: 
############################################################################

# Top 10 Movie Genres
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
movies=pd.read_csv("movies_updated.csv")
movies.head()

############################################################################
# Questions
# Find the top 10 genres with the largest number of movies.
movie_genre=movies["genre"].value_counts().head(10)
print(movie_genre)

############################################################################
# Create a vertical bar chart.
plt.figure(figsize=(10,6))
plt.bar(
    movie_genre.index,
    movie_genre.values,
)
plt.title('Top 10 Movie Genres',c="blueviolet" , size=20)
plt.xlabel('Genre' , size=16)
plt.ylabel('Number of Movies',c="orange" , size=16)
plt.xticks(rotation=45)
plt.show()

############################################################################
# Questions
# 1. Which genre contains the most movies?
most_movies=movie_genre.idxmax()
print(most_movies)

############################################################################
# 2. Which genre contains the fewest movies among the top 10?
fewest_movies=movie_genre.idxmin()
print(fewest_movies)

############################################################################
# 3. Why is a bar chart suitable for this task?
    # A bar chart is suitable because: 
        # Genres are categorical data.
        # We are comparing the number of movies across categories.
        # Bar lengths make differences easy to see.
        # It quickly identifies the most and least common genres.
        # It is one of the best visualizations for comparing category frequencies.
```

```python
# Task 4: Top 10 Directors

##########################################################################
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
movies=pd.read_csv("movies_updated.csv")

##########################################################################
# Find the top 10 directors with the largest number of movies.
top_director=movies["director"].value_counts().head(10)
print(top_director)

##########################################################################
# Create a horizontal bar chart.
plt.figure(figsize=(12,6))
plt.barh(
    top_director.index,
    top_director.values,
    edgecolor="black"
)
plt.title("Top 10 Directors",size=20)
plt.xlabel("Director", size=16)
plt.ylabel("Number of director" , size=16)
plt.show()

##########################################################################
# Questions
# 1. Which director produced the most movies?
most_movies=top_director.idxmax()
print("Most-Movie:" , most_movies)


##########################################################################
# 2. Why is a horizontal bar chart more suitable than a vertical bar chart for this visualization?
        # Horzontal bar is suoitable for this graph because :-
                 # 1. The title of the director line has very long so it is overlapped 
                 #     each other when draw each in vertical barchart
                 # 2. Makes ranking comparisons clearer.
```

```python
# Task 5: Distribution of Movie Ratings
#############################################################################

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
movies=pd.read_csv("movies_updated.csv")
# movies.head()

#############################################################################
# Using the score column, create a histogram.
plt.figure(figsize=(12,6))
plt.hist(
    movies["score"],
    bins=10,
    edgecolor="red"
)
plt.title("Distribution of Movie Ratings")
plt.xlabel("Movie Score")
plt.ylabel("Number of Movies")
plt.grid(alpha=1,color="green")
plt.show()

#############################################################################
# Questions
# 1. Which score range contains the largest number of movies?
score_ranges = pd.cut(movies["score"], bins=10)
score_ranges.value_counts().sort_values(ascending=False)

#############################################################################
# 2. Are most movies highly rated?
   # Most movies are average rated b/c:-
        # Scores above 8 are considered highly rated.
        # Scores between 5.5 and 7.5 are average to good.
        # No, most movies are moderately rated rather than highly rated

#############################################################################
# 3. Are low-rated movies common?
    # Low rated movies are uncommon b/c:-
        # There is a little amount of movies which is rated less than four
        # based on the above analysis

#############################################################################
# 4. What does the histogram tell you about movie ratings?
        # The most common movie are in average range.
        # very low-rated movies are rare.
        # highly rated movies are rare.
        # The overall shape of the rating distribution.
```

```python
# Task 6: Average Movie Score by Genre
#####################################################################################
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
movies=pd.read_csv("movies_updated.csv")
# movies.head()

#####################################################################################
# Calculate the average score for each genre.
avg_score_genre=(movies.groupby("genre")["score"]).mean().round(2)

#####################################################################################
# Sort the results from highest to lowest.
avg_score_genre=avg_score_genre.sort_values(ascending=False)
print(avg_score_genre)

#####################################################################################
# Create a bar chart.
plt.figure(figsize=(12,6))
avg_score_genre.plot(kind="bar", edgecolor="black")

plt.title("Average Movie Score by Genre", size=20)
plt.xlabel("Genre",size=16)
plt.ylabel("Average Score",size=16)

plt.xticks(rotation=45)
plt.show()


#####################################################################################
# Questions
# 1. Which genre has the highest average score?
hieghest_avg_score=avg_score_genre.idxmax()
print("Hieghest Average score is:",hieghest_avg_score)

#####################################################################################
# 2. Which genre has the lowest average score?
lowest_avg_score=avg_score_genre.idxmin()
print("Lowest Average Score is:",lowest_avg_score)

#####################################################################################
# 3. Are the most popular genres also the highest rated?
movies["genre"].value_counts()
    # No becuase:-
        # =>The most popular genre in the movies is commedy with value(1355), but
        # =>The hieghest rated movies is History with avearage rate of 8.3 wheare as 
        #   the commedy genre's are average rated 6.18

```
