```python
# Exercise 1: Removing Duplicate Records
import pandas as pd
import numpy as np
df = pd.DataFrame({
"student_id": [101, 102, 103, 102, 104, 101],
"name": ["Abel", "Sara", "John", "Sara", "Helen", "Abel"],
"score": [80, 90, 75, 90, 88, 80]
})

# Tasks

# 1. Identify duplicate rows.
df_duplicated=df.duplicated()
# print(df_duplicated)

# 2. Count duplicate rows.
print(df_duplicated.sum())

# 3. Remove duplicates.
df_cleaned = df.drop_duplicates()
# print(df_cleaned)

# 4. Keep only the last occurrence.
df_cleaned_first=df.drop_duplicates(keep="last")
# print(df_cleaned_first)

```

```python
# Exercise 5: Standardize Dates

# Dataset

df = pd.DataFrame({
"date":[
"2026-01-05",
"05/02/2026",
"March 10, 2026",
"2026.04.15"
]
})
# Tasks
# 1. Convert to datetime.
df["date"] = pd.to_datetime(df["date"],format="mixed")
# print(df["date"])
# print("=========================================")
# 2. Standardize to YYYY-MM-DD.
df["date"] = df["date"].dt.strftime("%Y-%m-%d")
# print(df["date"])
# print("=========================================")
# 3. Extract year.
df["date"]=pd.to_datetime(df["date"])
df["Year"]=df["date"].dt.year
# print(df["Year"])
# print("=========================================")
# 4. Extract month.
df["date"]=pd.to_datetime(df["date"])
df["Month"]=df["date"].dt.month
# print(df["Month"])

# print("====================================================")
# print(df)
```

```python
# Exercise 8: Handling Missing Values
# Dataset

import numpy as np

df = pd.DataFrame({
"name":["Abel","Sara",None,"John"],
"age":[20,np.nan,25,30]
})
# print(df)
# print("=======================")

# Tasks
# 1. Find missing values.
# print(df.isnull())
# print("=======================")
# 2. Fill missing ages with mean.
df["age"]=df["age"].fillna(df["age"].mean())
# print(df["age"])
# print("=======================")
# 3. Fill missing names with Unknown.
df["name"]=df["name"].fillna("Unkown")
# print(df["name"])
# print("=======================")
# print(df)
```

```python
# Exercise 11: Outlier Detection

# Dataset

df = pd.DataFrame({
"salary":[3500,4000,4200,3900,4100,45000]
})

# Tasks
# 1. Calculate Q1 and Q3.
Q1=df["salary"].quantile(0.25)
Q3=df["salary"].quantile(0.75)

# print(Q1,Q3)
# 2. Calculate IQR.
# print("To calculate IQR in from this project use differnce formula")
IQR=Q3-Q1
# print(IQR)

# 3. Identify outliers.
# print("================================================")
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR

# print(lower_bound)
# print(upper_bound)

outliers=[( df["salary"]<lower_bound)|(df["salary"]>upper_bound)]
# print(outliers)
# 4. Remove outliers.
cleaned_outliers=df[(df["salary"]>=lower_bound) & (df["salary"]<=upper_bound)]
# print(cleaned_outliers)
```

```python
# Exercise 12: Phone Number Standardization
# Dataset
df = pd.DataFrame({
"phone":[
"0911223344",
"+251911223344",
"251911223344",
"0911-223-344"
]
})

# Tasks
# 1. Remove symbols.
df["phone"]=df["phone"].str.replace("-","",regex=False)
df["phone"]=df["phone"].str.replace("+","",regex=False)

# 2. Convert all numbers to +251 format.
df["phone"]=df["phone"].apply(
    lambda x:
    "+251" + x[1:]
    if x.startswith("0")
    else
    "+251" + x[3:]
)

# print(df)
```

```python
# Exercise 13: Email Cleaning
# Dataset

df = pd.DataFrame({
"email":[
" ABEL@GMAIL.COM ",
"sara@gmail.com",
"JOHN@YAHOO.COM"
]
})
# Tasks
# 1. Remove spaces.
df["email"]=df["email"].str.strip()
# print(df)
# print("===================")
# 2. Convert to lowercase.
df["email_lower"]=df["email"].str.lower()
# print(df["email_lower"])
# 3. Validate emails.
# print("=====Email Validation=======")
df["email_valid"]=df["email"].str.contains("@")
# print(df["email_valid"])

# print(df)
```

```python
# Exercise 15: Mixed Data Cleaning Challenge
# Dataset

import numpy as np
import pandas as pd

df = pd.DataFrame({
"name":[" Abel ","Sara",None,"JOHN"],
"age":[20,np.nan,25,30],
"salary":[4000,4500,5000,45000],
"join_date":[
"2026-01-01",
"01/02/2026",
"March 10, 2026",
"2026.04.15"
]
})

# Tasks
# 1. Clean names.
df["name"]=df["name"].fillna("Unkown-Name")
print(df["name"])
# 2. Fill missing values.
df["age"]=df["age"].fillna(df["age"].mean())
print(df["age"])
# 3. Standardize dates.
df["join_date"]=pd.to_datetime(df["join_date"],format="mixed")
print(df["join_date"])
# 4. Detect outliers.
Q1 = df["salary"].quantile(0.25)
Q3 = df["salary"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
outliers = df[
(df["salary"] < lower) |
(df["salary"] > upper)
]

print(df)

print("Outliers")

print(outliers)
```
