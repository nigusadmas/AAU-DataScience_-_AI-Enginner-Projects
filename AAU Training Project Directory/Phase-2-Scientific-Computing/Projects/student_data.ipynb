```python
import pandas as pd 
```

```python
import numpy as np 
```

```python
students=pd.DataFrame({
    "name": ["Abel", "Hana", "Abel", None],
    "age": [21, np.nan, 21, 23],
    "department": ["CS", "IT", "CS", "SE"],
    "gpa": [3.5, 3.8, 3.5, np.nan]
    
})
```

```python
print(students)
```

```python
students.isnull()
```

```python
students.fillna("🚩 ")
```

```python
students["name"]=students["name"].fillna("Unkown")
print(students["name"])
```

```python
students["age"]=students["age"].fillna(students["age"].mode()[0])
print(students)
```

```python
students["gpa"]=students["gpa"].fillna(students["gpa"].mean())
print(students["gpa"])
```

```python
print("the output for cleaned data")
print(students)
```
