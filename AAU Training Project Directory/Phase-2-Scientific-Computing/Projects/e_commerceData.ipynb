```python
import pandas as pd
import numpy as np
import random

np.random.seed(42)

n = 50000

products = [
    "Laptop", "Phone", "Tablet", "Monitor",
    "Keyboard", "Mouse", "Printer", "Speaker"
]

categories = [
    "Electronics",
    "Accessories",
    "Office Equipment"
]

countries = [
    "Ethiopia",
    "Kenya",
    "Uganda",
    "Tanzania",
    "Rwanda"
]

cities = [
    "Addis Ababa",
    "Nairobi",
    "Kampala",
    "Dar es Salaam",
    "Kigali"
]

payment_methods = [
    "Cash",
    "Card",
    "Bank Transfer",
    "Mobile Money"
]

statuses = [
    "Completed",
    "Pending",
    "Cancelled"
]

df = pd.DataFrame({
    "Order_ID": np.arange(100001, 100001+n),

    "Customer_ID": np.random.randint(
        1000,
        5000,
        n
    ),

    "Customer_Name": [
        f"Customer_{i}"
        for i in range(n)
    ],

    "Email": [
        f"user{i}@gmail.com"
        for i in range(n)
    ],

    "Phone": [
        f"09{np.random.randint(10000000,99999999)}"
        for _ in range(n)
    ],

    "Country": np.random.choice(
        countries,
        n
    ),

    "City": np.random.choice(
        cities,
        n
    ),

    "Product": np.random.choice(
        products,
        n
    ),

    "Category": np.random.choice(
        categories,
        n
    ),

    "Quantity": np.random.randint(
        1,
        20,
        n
    ),

    "Unit_Price": np.random.randint(
        100,
        50000,
        n
    ),

    "Order_Date": pd.to_datetime(
        np.random.choice(
            pd.date_range(
                "2024-01-01",
                "2025-12-31"
            ),
            n
        )
    ),

    "Payment_Method": np.random.choice(
        payment_methods,
        n
    ),

    "Status": np.random.choice(
        statuses,
        n
    )
})

df["Total_Sales"] = (
    df["Quantity"] *
    df["Unit_Price"]
)

df.to_csv("e_commerce_data.csv",encoding="utf-8")
print("data saved succesfully")
```
