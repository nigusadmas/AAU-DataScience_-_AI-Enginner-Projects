```python
import pandas as pd

import numpy as np

university = pd.DataFrame({
    "student_id": [201, 202, 203, 204, 205, 206, 207, 208, 209, 210],
    "name": ["Abel", "Sara", None, "John", "Marta", None, "David", "Helen", "Tom", None],
    "department": ["CS", "IT", "CS", None, "SE", "IT", "SE", None, "CS", "IT"],
    "gpa": [3.5, 3.8, np.nan, 2.9, 3.7, np.nan, 3.2, 3.4, 3.9, np.nan],
    "scholarship": [5000, np.nan, 3000, 2000, np.nan, 4000, 3500, np.nan, 5000, np.nan]
})


print("=======Data cleaning Questions Answer=======")

# 1 count missing value
university.isnull().sum()

# 2 calculate missing value percentage
(university.isnull().sum()/len(university))*100

# 3 filling missing name with "Unkown"
university["name"]=university["name"].fillna("Unknown")


# 4 fill missing department by modal value of dep.t
university["department"]=university["department"].fillna(university["department"].mode()[0])


# 5 fill missing scholarship with zero(0)
university["gpa"]=university["gpa"].fillna(university["gpa"].median())

# 6 fill missing GPA by median value of Gpa
university["scholarship"]=university["scholarship"].fillna(0)
print(university)

print("\n")

print("=======Data Analysis Questions Answer=======")

# 7 count student by department
print("\n")
print("===Option 1===")
counts = {}

for index, row in university.iterrows():
    
    dept = row["department"]
    
    counts[dept] = counts.get(dept, 0) + 1

print(counts)

print("===Option 2===")
university["department"].value_counts()


# 8 counting scholarship reciepent
scholarship_reciepent=university["scholarship"].count()
print(scholarship_reciepent)
#This display 10 becuase scholarship data are filled with a number

# 9 find the hieghest gpa
Highest_Gpa=university["gpa"].max()
print(Highest_Gpa)

# 10 find the lowest gpa
Lowest_Gpa=university["gpa"].min()
print(Lowest_Gpa)

# 11 find Average gpa
Average_Gpa=university["gpa"].mean()
print(Average_Gpa)

# 12 find total scholarship amount
total_scholarship_sum=university["scholarship"].sum()
print(total_scholarship_sum)

```
