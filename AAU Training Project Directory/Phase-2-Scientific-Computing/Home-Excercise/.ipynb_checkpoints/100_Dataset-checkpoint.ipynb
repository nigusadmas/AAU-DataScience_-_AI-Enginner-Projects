```python
students["name"]=students["name"].str.strip().fillna("Unkown-Name")
# print(students["name"])
```

```python
students["age"]=pd.to_numeric(students["age"],errors="coerce")
# print(students["age"])
```

```python
students.loc[students["age"]>25,"age"]=np.nan
# print(students["age"])
```

```python
students.loc[students["age"]<15,"age"]=np.nan
# print(students["age"])
```

```python
students["age"]=students["age"].fillna(students["age"].mean())
# print(students["age"])
```

```python
students["age"]=students["age"].astype(int)
```

```python
# print(students["age"])
```

```python
students["gender"]=students["gender"].replace(
    {
        "male":"Male",
        "MALE":"Male",
        "female":"Female",
        "Femle":"Female",
        "Mle":"Male"
    }
)
```

```python
# print(students["gender"])
```

```python
students["gender"]=students["gender"].fillna(students["gender"].mode()[0])
# print(students["gender"])
```

```python
students["department"]=students["department"].replace(
    {
       "computer science":"Computer Science",
        "IT":"Information Technology",
        "Compter Science":"Computer Science",
        "software engineering":"Software Engineering",
        "Computer Scince":"Computer Science",
        "Software Eng":"Software Engineering"
    }
)
```

```python
students["department"]=students["department"].fillna(students["department"].mode()[0])
# print(students["department"])
```

```python
students.loc[students["gpa"]>4,"gpa"]=np.nan
students["gpa"]=students["gpa"].fillna(students["gpa"].mode()[0])
# print(students["gpa"])
```

```python
students.loc[students["gpa"]<0,"gpa"]=np.nan
students["gpa"]=students["gpa"].fillna(students["gpa"].mode()[0])
# print(students["gpa"])
```

```python
students["scholarship"]=pd.to_numeric(students["scholarship"],errors="coerce")
# print(students["scholarship"])

students["scholarship"]=students["scholarship"].fillna(students["scholarship"].mean())
# print(students["scholarship"])
```
