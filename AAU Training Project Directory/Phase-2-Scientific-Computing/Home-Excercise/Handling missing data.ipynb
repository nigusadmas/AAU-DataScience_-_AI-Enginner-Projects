```python
# Create the dataset
data = {'ID': ['Eth - 101', 'Eth – 102', 'Eth - 103', 'Eth - 104', 'Eth - 105'],
'Name': [' Haileslassie Zeferu Abel', ' Worku Abenezer Girma', ' Abenezer Reta Tamirat', ' Abraham Melisew Fireda', ' Hailu Abraham Niguse'],
'Age': [25, None, 29, 33, None],
'Salary': [55000, 48000, None, 62000, None]}
```

```python
print(df)
```

```python
# check duplicates
df.duplicated()
```

```python
# if there is duplicate in the dataset you can remove like this
df.drop_duplicates()
```

```python
# Remove rows with missing values
df_cleaned = df.dropna()
print(df_cleaned)
```

```python
# Method 3: Flagging
import pandas as pd
# Create a column-based flag
for column in df.columns:
    if df[column].isnull().any():
        df[f'{column}_Flag'] = df[column].isna().apply(lambda x: 'Missing'if x else 'Not Missing')
# Create a row-based flag
df['Missing_Flag'] = df.isnull().any(axis=1).apply(lambda x: 'Missing' if x else 'Not Missing')
```

```python
# Removing_Duplicates
import pandas as pd
# Sample dataset
data = {'ID': ['Eth - 101', 'Eth - 102', 'Eth - 103', 'Eth - 104', 'Eth - 105', 'Eth - 102', 'Eth - 104', 'Eth - 106'],
'Name': ['Haileslassie Zeferu Abel', 'Worku Abenezer Girma', 'Abenezer Reta Tamirat', 'Abraham Melisew Fireda', 'Hailu Abraham Niguse', 'Worku Abenezer Girma', 'Abraham Melisew Fireda', 'Hailu Abraham Niguse'],
'Age': [25, 29, 29, 33, 29, 29, 33, 29],
'Salary': [55000, 48000, 55000, 62000, 55000, 48000, 62000, 55000],
'City': ['Addis Ababa', 'Adama', 'Addis Ababa', 'Adama', 'Addis Ababa', 'Adama', 'Adama', 'Addis Ababa']}
# Original Dataset
df = pd.DataFrame(data)
print("Original Dataset")
print(df, '\n')
```

```python
df_cleaned=df.drop_duplicates()
```

```python
print("Duplicates are Removed based on the exact Match")
print(df_cleaned, '\n')
```

```python
# Partial Match - based on specific columns only (e.g., 'ID')
df_cleaned_id = df.drop_duplicates(subset=['ID'])
print("Duplicates Removed - Partial Match based on ID")
print(df_cleaned_id, '\n')
```

```python
# Partial Match - based on specific columns only (e.g., 'ID')
df_cleaned_name = df.drop_duplicates(subset=['Name'])
print("Duplicates Removed - Partial Match based on Name")
print(df_cleaned_name, '\n')
```

```python
# Partial Match - based on specific columns only (e.g., 'ID' and 'Name')
df_cleaned_specific = df.drop_duplicates(subset=['ID', 'Name'])
print("Duplicates Removed - Partial Match based on ID and Name")
print(df_cleaned_specific, '\n')
```
