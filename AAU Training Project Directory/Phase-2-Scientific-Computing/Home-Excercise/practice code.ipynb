```python
import pandas as pd
import numpy as np

students = pd.DataFrame({
    "student_id": [101,102,103,104,105,106,107,108,108],
    "name": [
        "  Abel ",
        "Hana",
        "Mekdes",
        "Solomon",
        None,
        "dawit",
        "Sara ",
        "Yonas",
        "Yonas"
    ],
    "age": [
        21,
        22,
        np.nan,
        25,
        20,
        "twenty one",
        23,
        200,
        200
    ],
    "department": [
        "Computer Science",
        "computer science",
        "IT",
        "Information Technology",
        "CS",
        "cs",
        "Software Engineering",
        "SE",
        "SE"
    ],
    "gpa": [
        3.5,
        3.8,
        3.2,
        np.nan,
        4.9,
        3.6,
        3.7,
        2.9,
        2.9
    ],
    "scholarship": [
        "Yes",
        "No",
        "yes",
        "YES",
        None,
        "no",
        "Yes",
        "No",
        "No"
    ]
})
print(students)
```

```python
students.fillna("🚩")
```

```python
students["name"]=students["name"].str.strip().fillna("Unkoun_Name")
print(students["name"])
```

```python
students["age"] = pd.to_numeric(students["age"], errors="coerce")
print(students["age"])
```

```python
students["age"] = students["age"].replace({
    200:20
})
```

```python
print(students["age"])
```

```python
students["age"]=students["age"].fillna(students["age"].mean())
students["age"]=students["age"].astype(int)
print(students["age"])
```

```python
print(students)
```

```python
students["department"]=students["department"].replace({
    "Computer Science":"CS",
    "computer science":"CS",
    "Information Technology":"IT",
    "cs":"CS",
    "Software Engineering":"SE"
    
    
})
```

```python
print(students["department"])
```

```python
students["scholarship"].fillna(students["scholarship"].mode()[0])
students["scholarship"].str.title()
print(students["scholarship"])
```

```python
students["scholarship"]=students["scholarship"].replace(
    {
        "YES":"Yes",
        "yes":"Yes",
        "no":"No"
        
    }
)
print(students["scholarship"])
```
