```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
freelance_gigs = {
    "Service": ["Python", "Web Design", "App Development", "AWS"],
    "Price_USD": [200, 150, 300, 98]
}
```

```python
Price_df = pd.DataFrame(freelance_gigs)
# print(Price_df)
```

```python
plt.bar(Price_df["Service"],Price_df["Price_USD"] , color="blue")
plt.title("Freelance Service Prices")
plt.xlabel("Services in A company")
plt.ylabel("Price Of Services")

plt.show()
```
