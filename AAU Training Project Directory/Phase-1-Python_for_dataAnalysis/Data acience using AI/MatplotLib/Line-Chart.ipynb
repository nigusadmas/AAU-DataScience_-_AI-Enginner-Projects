```python
import matplotlib.pyplot as plt

# 1. ዳታውን እናዘጋጃለን (ወራት እና የገቢ መጠን)
months = ["Jan", "Feb", "Mar", "Apr", "May"]
income = [120, 180, 150, 270, 310]

# 2. ላይን ግራፍ ለመስራት plt.plot() እንጠቀማለን
# marker='o' ማለት እያንዳንዱ ቁጥር ላይ ክብ ነጠብጣብ አድርግ ማለት ነው
# linestyle='--' ማለት መስመሩን የተቆራረጠ አድርገው ማለት ነው
plt.plot(months, income, color="blue", marker='o', linestyle='-')

# 3. ርዕስ እና መግለጫ
plt.title("Freelance Income Growth (Jan - May)")
plt.xlabel("Months")
plt.ylabel("Income in USD")

# 4. ግራፉን ማሳየት
plt.show()
```
