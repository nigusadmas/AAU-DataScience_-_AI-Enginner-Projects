```python
#scrap data(Qoutes and authors from "Quotes to scrap" 
# website and convert into pandas DataFrame )

import pandas as pd
import requests
from bs4 import BeautifulSoup

url="https://quotes.toscrape.com/"

response=requests.get(url)

soup=BeautifulSoup(response.text , "html.parser")

Qoute_tag=soup.find_all("span" , class_="text")
author_Name=soup.find_all("small" , class_="author")
# print(Qoute_tag)
# print(author_Name)

Qoute_list=[]
Author_List=[]

for tag in Qoute_tag:
    Qoute_list.append(tag.text)
    # print(Qoute_list)
for author in author_Name:
    Author_List.append(author.text)
    # print(Author_List)

data={"Qoutes":Qoute_list,
      "author":Author_List
     }
# print(data)

df=pd.DataFrame(data )


print(df)

```
