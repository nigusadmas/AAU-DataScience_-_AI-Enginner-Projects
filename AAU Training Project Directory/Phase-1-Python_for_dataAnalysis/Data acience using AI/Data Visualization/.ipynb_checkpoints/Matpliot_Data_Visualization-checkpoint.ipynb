```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
freelance_gigs = {
    "Service": ["Python", "Web Design", "App Development", "AWS"],
    "Price_USD": [200, 150, 300, 98]
}
```

```python
Price_df = pd.DataFrame(freelance_gigs)
```

```python
# 2. የባር ቻርት እንሰራለን (X-axis = Service, Y-axis = Price_USD)
```

```python
# 3. ለግራፉ ርዕስ እና መግለጫ እንሰጣለን
plt.bar(Price_df["Service"], Price_df["Price_USD"], color="skyblue")
plt.title("Freelance Service Prices")
plt.xlabel("Services")
plt.ylabel("Price in USD")
# 4. ግራፉን በስክሪን ላይ እንዲታይ እናደርጋለን
plt.show()
```
