```python
import pandas as pd

freelance_gigs={
    "Service":["Python","Web Design", "App Development","AWs"],
    "Price_USD":[200,150, 300,98]
}
Price_df=pd.DataFrame(freelance_gigs)

Price_USD=Price_df["Price_USD"]
print(Price_USD)

premium_USDTT=Price_df[Price_USD>100]
print(premium_USDTT)

premium_gigs=premium_USDTT["Service"]


print(premium_gigs)

```
