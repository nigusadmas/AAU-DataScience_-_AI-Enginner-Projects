```python
import requests
from bs4 import BeautifulSoup
import time

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
}

# 1. መጀመሪያ ከዋናው ገጽ ላይ ሊንኮችን እንሰበስባለን
main_url = "https://newbusinessethiopia.com/"
response = requests.get(main_url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")

articles = soup.find_all("article")
links_list = []

for article in articles:
    link_tag = article.find("a")
    if link_tag and link_tag["href"] not in links_list:
        links_list.append(link_tag["href"])

# 2. የ Deep Scraping ክፍል (የመጀመሪያዎቹን 2 ሊንኮች ብቻ እንፈትሽ)
test_links = links_list[:2]
full_stories = []

print("--- Deep Scraping ተጀምሯል ---")
for link in test_links:
    print(f"ወደ ውስጥ በመግባት ላይ፦ {link}")
    
    # ወደ እያንዳንዱ የዜና ገጽ ቀጥታ እንገባለን
    inside_response = requests.get(link, headers=headers)
    inside_soup = BeautifulSoup(inside_response.text, "html.parser")
    
    # በዚህ ድረ-ገጽ ላይ ሙሉ የዜናው ይዘት አብዛኛውን ጊዜ በ <div class="entry-content"> ውስጥ ይገኛል
    content_div = inside_soup.find("div", class_="entry-content")
    
    if content_div:
        # በውስጡ ያሉትን ሁሉንም የፓራግራፍ (<p>) ፅሁፎች በአንድ ላይ እናያይዛቸዋለን
        paragraphs = content_div.find_all("p")
        full_text = "".join([p.text for p in paragraphs])
        full_stories.append(full_text.strip())
    else:
        full_stories.append("የዜናው ዝርዝር አልተገኘም")
        
    time.sleep(20) # እረፍት መስጠት እንዳንረሳ!

# ውጤቱን ማየት
for i, story in enumerate(full_stories):
    print(f"\n📰 ዜና {i+1} ሙሉ ይዘት (የመጀመሪያዎቹ 200 ቃላት)፦")
    print(story[:200] + "...")
```
