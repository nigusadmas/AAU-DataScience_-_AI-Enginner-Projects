```python
import pandas as pd
```

```python
import requests
```

```python
import time
```

```python
from bs4 import BeautifulSoup
```

```python
 headers = {
        "User-Agent": (
            "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
            "AppleWebKit/537.36 (KHTML, like Gecko) "
            "Chrome/137.0.0.0 Safari/537.36"
        ),
        "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8",
        "Accept-Language": "en-US,en;q=0.5",
            }
```

```python
title_list=[]
image_list=[]
date_list=[]
para_list=[]
```

```python
for page in range(1,6):
    url=f"https://newbusinessethiopia.com/topics/oil-gas/page/{page}"
    print(f"The page is running on the Link:{url}")
    response=requests.get(url , headers=headers)
    soup=BeautifulSoup(response.content ,"html.parser" )
    articles=soup.find_all("article")

    
    for article in articles:
        title = article.find('div').find('h2').find('a')
        image=article.find("img")
        date=article.find("p" , class_="nbe-single-meta")
        paragraph=article.find("div" , class_="nbe-archive-text")

        # check wether the data is in the page or not if it is append into the empty List
        if title and image and date and paragraph:
            date_list.append(date.text.strip())
            image_list.append(image["src"])
            title_list.append(title.text.strip())
            para_list.append(paragraph.text.strip())

    
    time.sleep(10)
    
df=pd.DataFrame({
    "Title-List":title_list,
    "Image-List":image_list,
    "Date-List":date_list,
    "Para-List":para_list
})

print(df)
# convert to csv file
df.to_csv("Oil_Gas_Data.csv",encoding="utf-8")
print("CSV file created successfully")
```
