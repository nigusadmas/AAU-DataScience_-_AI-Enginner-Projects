```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://newbusinessethiopia.com/"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")

articles = soup.find_all("article")

titles_list = []
categories_list = []

for article in articles:
    title_tag = article.find("h3")
    
    if title_tag:
        titles_list.append(title_tag.text.strip())
        
        # 💡 የክላስ ስሞችን ዝርዝር በሊስት መልክ እንቀበላለን
        class_list = article.get("class", [])
        
        # "category-" የሚለውን ቃል ፈልገን የዘርፉን ስም ብቻ እንለያለን
        found_category = "General" # ክላሱ ካልተገኘ General እንዲል
        for c in class_list:
            if c.startswith("category-"):
                found_category = c.replace("category-", "").capitalize()
                break
                
        categories_list.append(found_category)

# በፓንዳስ ማደራጀት
df = pd.DataFrame({
    "News_Title": titles_list,
    "Category": categories_list
})

print(df)
print("###############################################")
print(df["Category"].value_counts())
```
