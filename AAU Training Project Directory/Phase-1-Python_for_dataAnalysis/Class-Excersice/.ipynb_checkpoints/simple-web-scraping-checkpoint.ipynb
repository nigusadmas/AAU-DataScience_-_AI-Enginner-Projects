```python
from bs4 import BeautifulSoup
import pandas as pd

html = """
<div class="container">
    <h1 class="title">Algorithm</h1>
    <h1 class="title">Python</h1>
    <h1 class="title">JavaScript</h1>
    <h1 class="title">Java</h1>
"""

soup = BeautifulSoup(html, "html.parser")

headings = soup.find_all("h1", class_="title")

data = []

for i, item in enumerate(headings, start=1):
    text = item.text.strip()
    data.append({
        "id": i,
       "item": text
    })

for row in data:
    print(row)


```
