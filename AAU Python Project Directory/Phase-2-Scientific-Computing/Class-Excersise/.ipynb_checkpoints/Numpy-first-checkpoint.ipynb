```python
matrix_1d = np.array([1, 2, 3])
```

```python
matrix_2d = np.array([[1, 2], [3, 4]])
```

```python
matrix_3d=np.array([[[1,2,4,6],[2,4,6,8],[4,6,9,13]]])
```

```python
# Fast element-wise multiplication (Vectorization)
```

```python
result = matrix_1d * 2
```

```python
print(result)  # Output: [2, 4, 6]
print("The Dimension for this matrix is:",result.shape ,"Dimensin")
```
