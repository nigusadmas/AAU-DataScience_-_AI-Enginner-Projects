```python
import numpy as np
a = np.array(
    [
        [1, 2], [3, 4]
    ]
            )
b = np.array(
    [
        [4, 3], [2, 1]
    ]
)
print ("Adding 1 to every element:", a + 1)
print ("\nSubtracting 2 from each element:", b - 2)
print ("\nSum of all array elements in a: ", a.sum())
print ("\nSum of all array elements in b: ", b.sum())
print ("\nArray sum:\n", a + b)
```
