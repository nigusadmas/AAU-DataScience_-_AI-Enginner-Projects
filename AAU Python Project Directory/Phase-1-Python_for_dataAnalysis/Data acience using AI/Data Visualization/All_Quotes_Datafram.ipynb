```python
import requests
from bs4 import BeautifulSoup
import pandas as pd 
```

```python
url="https://quotes.toscrape.com/"
```

```python
response=requests.get(url)
```

```python
soup=BeautifulSoup(response.text,"html.parser")
```

```python
# print(soup)
```

```python
all_tag_quote=soup.find_all("span" , class_="text")
```

```python
# print(all_tag_quote)
```

```python
tag_qoute_list=[]
```

```python
for tag in all_tag_quote:
    tag_qoute_list.append(tag.text)
    # print(tag_qoute_list)
    
```

```python
data={"Quotes":tag_qoute_list}
```

```python
df=pd.DataFrame(data)
```
