```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url="https://quotes.toscrape.com/"
```

```python
response=requests.get(url)
```

```python
soup=BeautifulSoup(response.text , "html.parser")
```

```python
Qoute_tag=soup.find_all("span" , class_="text")
author_Name=soup.find_all("small" , class_="author")
# print(Qoute_tag)
# print(author_Name)
```

```python
Qoute_list=[]
Author_List=[]
```

```python
for tag in Qoute_tag:
    Qoute_list.append(tag.text)
    # print(Qoute_list)
for author in author_Name:
    Author_List.append(author.text)
    # print(Author_List)
```

```python
data={"Qoutes":Qoute_list}
# print(data)

author_data={"author":Author_List}
print(author_data)
```

```python
df=pd.DataFrame(data)
```
