```python
from bs4 import BeautifulSoup

# 1. አንድ የድረ-ገጽ HTML ኮድ በፅሁፍ መልክ አዘጋጀን
html_content = """
<html>
    <body>
        <h1 id="title">Nigus Freelance Portal</h1>
        <p class="description">Find the best tech gigs here.</p>
    </body>
</html>
"""

# 2. BeautifulSoup ን ተጠቅመን HTML ኮዱን እንዲመረምር እናዘጋጀዋለን
soup = BeautifulSoup(html_content, "html.parser")

# 3. የ <h1> ታግ ፅሁፉን ብቻ ለይተን እናውጣ
main_heading = soup.find("h1").text
print(main_heading)
```
