```python
import pandas as pd

freelance_gigs={
    "Service":["Web Design", "App Development"],
    "Price_USD":[150, 300]
}

Price_df=pd.DataFrame(freelance_gigs)
print(Price_df)
```

```python
# To print service column only
Services=Price_df["Service"]
print(Services)
```

```python
Price_USDT=Price_df["Price_USD"]
print(Price_USDT)
```

```python
App_Dev=Price_df[Price_USDT>200]
print(App_Dev["Service"])
```
