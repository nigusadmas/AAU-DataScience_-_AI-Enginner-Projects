```python
    headers = {
        "User-Agent": (
            "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
            "AppleWebKit/537.36 (KHTML, like Gecko) "
            "Chrome/137.0.0.0 Safari/537.36"
        ),
        "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8",
        "Accept-Language": "en-US,en;q=0.5",
            }
```

```python
image_list=[]
title_list=[]
para_list=[]
date_list=[]
# catagory_list=[]
```

```python
home_lin=["investment" , "page"]
for page in range (1,4):
    url=f"https://newbusinessethiopia.com/topics/investment/page/{page}"
    print(f"link Download progress:{url}")
    response=requests.get(url,headers=headers)
    soup=BeautifulSoup(response.text,"html.parser")
    articles = soup.find_all("article")
    # All_catagory=soup.find_all("ul" , class_="nbe-primary-menu")
    # for catagory in All_catagory:
    #     categories=catagory.find_all("li")
    #     for li in categories:  # Loop through each li element
    #         catagory_list.append(li.text.strip())
    for article in articles:
        image = article.find("img")
        title = article.find('div').find('h2').find('a')
        date=article.find("p",class_="nbe-single-meta")
        paragraph=article.find("div", class_="nbe-archive-text")
        if image and title and date and paragraph:
            date_list.append(date.text.strip())
            image_list.append(image["src"])
            title_list.append(title.text.strip())
            para_list.append(paragraph.text.strip())

    time.sleep(25)
    print("===page scraped with delay time 20 second between each pages===")

# convert this data into dataframes
df=pd.DataFrame({
    # "catagory_list":catagory_list,
    "date_list":date_list,
    "image_list":image_list,
    "title_list":title_list,
    "para_list":para_list  
})

print(df)

# df.to_csv("All_Investment_Page.csv",index=False)
print("Csv file created Successfully")
```
