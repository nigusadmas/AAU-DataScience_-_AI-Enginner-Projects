```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time  # ⬅️ አዲስ ነገር! ድረ-ገጹ ላይ ጫና እንዳናበዛ በየገጹ መሀል እረፍት ለመስጠት

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
}

# ሁሉንም መረጃዎች የምንሰበስብባቸው ዋና ሊስቶች
all_titles = []
all_links = []
all_images = []

# ከገጽ 1 እስከ ገጽ 3 ድረስ እንዞራለን (range(1, 4) ማለት 1, 2, 3 ማለት ነው)
for page in range(1, 4):
    # በየዙሩ የገጹን ቁጥር በ URL ውስጥ እንቀይራለን
    url = f"https://newbusinessethiopia.com/page/{page}/"
    
    print(f"በማውረድ ላይ፦ {url} ...")
    
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, "html.parser")
    # print(soup)
    
    articles = soup.find_all("article")
    # print(articles)
    for article in articles:
        title_tag = article.find("h3")
        link_tag = article.find("a")
        img_tag = article.find("img")
        
        if title_tag and link_tag and img_tag:
            all_titles.append(title_tag.text.strip())
            all_links.append(link_tag["href"])
            all_images.append(img_tag["src"])
            
    # 🛑 Pro-Tip: እውነተኛ ሰውን ለመምሰል እና ድረ-ገጹን ላለማጨናነቅ በየገጹ መሀል 2 ሰከንድ እንጠብቃለን
    time.sleep(20)

# ሁሉንም ገጾች አዳርሰን ስንጨርስ ወደ Pandas DataFrame እንቀይራለን
df = pd.DataFrame({
    "News_Title": all_titles,
    "Article_Link": all_links,
    "Image_URL": all_images
})

print("\n--- የመጨረሻው የተሰበሰበው ዳታ ማጠቃለያ ---")
print(f"በአጠቃላይ ከ 3 ገጾች የተሰበሰቡ የዜናዎች ብዛት፦ {len(df)}")
print(df)  # የመጀመሪያዎቹን ዜናዎች ያሳያል
```
